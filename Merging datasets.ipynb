```python
import pandas as pd
```

```python
train_claim = pd.read_csv('data/TrainData/Train_Claim.csv')
train_demographics = pd.read_csv('data/TrainData/Train_Demographics.csv')
train_policy = pd.read_csv('data/TrainData/Train_Policy.csv')
train_vehicle = pd.read_csv('data/TrainData/Train_Vehicle.csv')
train_data = pd.read_csv('data/TrainData/Traindata_with_Target.csv')
```

```python
train_vehicle
```

```python
veh_table = pd.pivot_table(train_vehicle, values='VehicleAttributeDetails', index=['CustomerID'],
                       columns=['VehicleAttribute'], aggfunc=lambda x: ' '.join(x))
veh_table
```

```python
df1 = pd.merge(train_data,train_claim, on = ['CustomerID'], how='inner')
df2 = pd.merge(df1,train_demographics,on = ['CustomerID'], how='inner')
df3 = pd.merge(df2,train_policy,on = ['CustomerID'], how='inner')
df = pd.merge(df3,veh_table,on = ['CustomerID'], how='inner')
df
```

```python
df.to_csv('data/india_fraud.csv', index=False)
```
